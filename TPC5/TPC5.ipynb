```python
def criarTurma():
    turma=[]
    return turma


def inserirAluno(listat):
    a=input("Digitar o nome do aluno:")
    i=int(input("Digitar o id desse aluno:"))
    n1=int(input("Digitar a nota que obteve nos TPCs:"))
    n2=int(input("Digitar a nota que obteve no Projeto:"))
    n3=int(input("Digitar a nota que obteve no Teste:"))
    notas=[n1,n2,n3]
    al=(a,i,notas)
    listat.append(al)
    print(listat)
    return listat


def listarTurma(listat):
    print(f"Lista dos alunos na Turma {listat}:")
    for t in listat:
        print(f"O aluno {t[0]}, tem o id {t[1]}, obteve a nota {t[2][0]} no TPC, no projeto obteve {t[2][1]} e no teste conseguiu {t[2][2]}")
        return
    
def idAluno(listat,i):
    for t in listat:
        if i == t[1]:
            print(f"O aluno com id {i} chama-se {t}.")
        else:
            print("Não foi encontrado o aluno que procura")
    return


def guardarTurma(listat,fnome):
    file= open(fnome,"w") 
    result= ""
    for c in listat:
        al, i, notas= c
        result+= str(al) + ";" + str(i) + ";" + str(notas[0]) + "::" + str(notas[1]) + "::" + str(notas[2]) + "|"
        result+="\n"
    file.write(result)
    file.close()


def recuperarTurma(fnome):
    listat=[]
    file= open(fnome, "r")
    text=file.read()
    turma_text=text.split("\n")
    
    
    for p_text in turma_text[:-1]:
        t=[]
        t_text=p_text.split("|")


        for t_text in t_text[:-1]:
            al,i,notas= t_text.split(";")


            notas_Lista=[]
            for n in notas.split("::"):
                notas_Lista.append(float(n))
            alun=(str(al),int(i),list(notas_Lista))
            t.append(alun)
        listat.append(t)
    file.close()
    return listat

def menu():
    print("\n====menu====")
    print("1 - Criar uma turma")
    print("2 - Inserir um aluno na turma")
    print("3 - Listar a turma")
    print("4 - Consultar um aluno por id")
    print("5 - Guardar a turma em ficheiro")
    print("6 - Carregar uma turma dum ficheiro")
    print("0 - Sair da aplicação")
    print("=============")

modo=-1
menu()
while modo!=0:
    modo=int(input("O que pretende fazer?"))
    if modo==0:
        print("Saiu")
    elif modo==1:
        turma=criarTurma()
    elif modo==2:
        inserirAluno(turma)
    elif modo==3:
        listarTurma(turma)
    elif modo==4:
        iden=int(input("Qual é o id que procura?:"))
        idAluno(turma,iden)
    elif modo==5:
        print(f"A sua turma foi guardada como: turma")
        guardarTurma(turma,"turma.txt")
    elif modo==6:
        print(recuperarTurma("turma.txt"))
    else:
        print("Tente outra vez, escolha outro número de 0 a 6")
    



```

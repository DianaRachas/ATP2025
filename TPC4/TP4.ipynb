### TPC4: Aplicação para Gerir um Cinema

Suponha que está a desenvolver uma aplicacão para gestão de um conjunto de salas de cinema de um centro comercial. 
Nesse centro comercial existem algumas salas de cinema (que poderão estar a exibir filmes ou não), cada sala tem uma determinada 
lotação, uma lista com a referência dos bilhetes vendidos (lugares ocupados; cada lugar é identificado por um número inteiro), e cada sala tem um filme associado.

Considera a seguinte sugestão para o modelo dos cinemas:
```
Cinema = [Sala]
Sala = [nlugares, Vendidos, filme]
nlugares = Int
filme = String 
Vendidos = [Int]
```
  
Que poderá ser usado num programa da seguinte forma:
```
sala1 = (150, [], "Twilight")
sala2 = (200, [], "Hannibal")
cinema1 = []
...
cinema1 = inserirSala(cinema1,sala1)
cinema1 = inserirSala(cinema1,sala2)
...
listar(cinema1)
...

if(disponivel(cinema1, "Twilight", 17 )):
  cinema1 = vendebilhete(cinema1, "Twilight", 17 )
...
listardisponibilidades(cinema1)
...
```

Especifique as funções utilizadas no exemplo:

1. `listar( cinema )` - que lista no monitor todos os filmes que estão em exibição nas salas do cinema passado como argumento;
2. `disponivel( cinema, filme, lugar )` - que dá como resultado **False** se o lugar lugar já estiver ocupado na sala onde o filme está a ser exibido e dará como resultado **True** se o inverso acontecer;
3. `vendebilhete( cinema, filme, lugar )` - que dá como resultado um novo cinema resultante de acrescentar o lugar à lista dos lugares ocupados, na sala onde está a ser exibido o filme;
4. `listardisponibilidades( cinema )` - que, para um dado cinema, lista no monitor para cada sala, o filme que está a ser exibido e o total de lugares disponíveis nessa sala (número de lugares na sala menos o número de lugares ocupados);
5. `inserirSala( cinema, sala )` - que acrescenta uma sala nova a um cinema (devendo verificar se a sala já existe);
6. Acrescente todas as outras funcionalidades que achar necessárias;
7. À semelhança do exercício 3, construa uma aplicação com um menu de interface para as operações.


````python
def existeCinema(listac,nomec):
    cond=False
    for c in listac:                                      #como ir a cada sala?
        if c[2] == nomec:
            cond = True
    return cond
    
def inserirCinema(listac, c):
    if not existeCinema(listac,c[2]):
        listac.append(c)
    return listac

def listarc(listac):
    for c in listac:
        print(f"A sala {c[2]} tem {c[0]} lugares e {len(c[1])} são os lugares ocupados")
    return

def disponivel( listac, nomec, lugar):
    cond=True
    for c in listac:
        if c[2] == nomec:
            if lugar in c[1]:
                cond= False
    print(cond)
    return cond

def vendabilhetes( listac, nomec, lugar):
    novalista=[]
    for c in listac:
        if c[2] ==nomec:
            if disponível(listac, nomec, lugar): 
                novalista.append([c[0], c[1] + [lugar],c[2]])
            else:
                novalista.append([c[0], c[1],c[2]])
        else:
            novalista.append([c[0], c[1],c[2]])
    listac=novalista
    print(listac)
    return novalista

def listardisponibilidades(listac):
    for c in listac:
        disponibilidade= c[0]- len(c[1])
        print(f"{c[2]} tem estes lugares ocupados:{c[1]}")
        print(f"{c[2]} tem, exatamente, {disponibilidade} estes lugares disponíveis")
    return

def listarcc(listac, nomec):
    for c in listac:
        if c[2]==nomec:
            print(f"O cinema {c[2]} tem {c[0]} lugares e estes lugares estão ocupados:{c[1]}")
    return


def libertaLugar(listac, nomec, lugar):
    novalista=[]
    for c in listac:
        if c[2] == nomec:            
            if lugar in c[1]:
                c[1].remove(lugar)
            novalista.append([c[0], c[1],c[2]])
        else:
            novalista.append([c[0], c[1],c[2]])
    else:
        novalista.append([c[0], c[1],c[2]])
    print(novalista)
    return novalista

def criaCinema(listac, nomec, lugares):
    c=[lugares,[],nomec]
    if not existeCinema(listac,c):
        listac.append(c)
    print(listac)
    return listac
    
def removeCinema(listac, nomec):
    for c in listac:
        if c[2]==nomec:
            if c[1]==[]:
                listac.remove(c)
    print(listac)
    return listac

modo=-1
cinema1=[]  #cinema tem várias salas

while modo!=0:
    modo = int(input("""Qual operação quer executar: 
                   (1)--> inserirCinema; 
                   (2)--> listar cinema; 
                   (3)--> ver se o cinema tem lugares disponiveis; 
                   (4)--> venda de bilhetes;
                   (5)--> mostra de quais lugares estão disponiveis;
                   (6)--> uma lista de quais lugares estão ocupados no cinema;
                   (7)--> liberta lugar do cinema;
                   (8)--> cria um cinema ;
                   (9)--> remove um cinema se não tiver pessoas para assistir;
                   (0)--> Sair
                   --> """))
    
    if modo==0:
        print("Sair")
    elif modo==1:
        nome = input("Cinema:")
        lug = int(input("Lugares:"))
        sala = [lug, [], nome]
        inserirCinema(cinema1,sala)
    elif modo==2:
        listarc(cinema1)
    elif modo==3:
        sala = input("Cinema:")
        lug = int(input("Lugar:"))
        disponivel(cinema1, sala, lug)
    elif modo==4:
        sala = input("Cinema:")
        lug = int(input("Lugar:"))
        vendabilhetes(cinema1, sala, lug)
    elif modo==5:
        listardisponibilidades(cinema1)
    elif modo==6:
        sala = input("Cinema:")
        listarcc(cinema1, sala)
    elif modo==7:
        sala = input("Cinema:")
        lug = int(input("Lugar:"))
        libertaLugar(cinema1, sala, lug)
    elif modo==8:
        sala = input("Cinema:")
        lug = int(input("Lugar:"))
        criaCinema(cinema1, sala, lug)
    elif modo==9:
        sala = input("Cinema:")
        removeCinema(cinema1, sala)
    else:
        print("Volta a tentar:")
````
